Caricamento dei dataframes

```python
#Loading the datasets
df_2122 = pd.read_csv(filename_1, sep=",")
df_2223 = pd.read_csv(filename_2, sep=",")
```

```python
display_all(df_2223)
```

Calcoliamo la classifica con tutti gli altri parametri a partire dai risultati delle partite.

```python
# Creo il set delle squadre del campionato
Teams = set(df_2223["HomeTeam"])

stats = dict()

# Per ogni squadra calcolo i seguenti dati
for x in Teams:
    
    stats[x] = dict()
    
    stats[x]["goal_subiti"]      = sum(df_2223[df_2223["HomeTeam"] == x].FTAG) + sum(df_2223[df_2223["AwayTeam"] == x].FTHG)
    stats[x]["vittorie"]         = len(df_2223[(df_2223.HomeTeam == x) & (df_2223.FTHG > df_2223.FTAG)]) + len(df_2223[(df_2223.AwayTeam == x) & (df_2223.FTAG > df_2223.FTHG)])
    stats[x]["pareggi"]          = len(df_2223[(df_2223.HomeTeam == x) & (df_2223.FTHG == df_2223.FTAG)]) + len(df_2223[(df_2223.AwayTeam == x) & (df_2223.FTAG == df_2223.FTHG)])
    stats[x]["sconfitte"]        = len(df_2223[(df_2223.HomeTeam == x) & (df_2223.FTHG < df_2223.FTAG)]) + len(df_2223[(df_2223.AwayTeam == x) & (df_2223.FTAG < df_2223.FTHG)])
    stats[x]["goal_fatti"]       = sum(df_2223[df_2223["HomeTeam"] == x].FTHG) + sum(df_2223[df_2223["AwayTeam"] == x].FTAG)
    stats[x]["differenza_reti"]  = stats[x]["goal_fatti"] - stats[x]["goal_subiti"]
    stats[x]["punti"]            = stats[x]["vittorie"] * 3 + stats[x]["pareggi"]
```

## Classifica punti

```python
# Trasformo il dizionario ottenuto in un dataframe e lo ordino per punti
statistiche = pd.DataFrame.from_dict(stats).transpose().sort_values("punti", ascending=True)

#Insert lables
for index, value in enumerate(statistiche.punti):
    plt.text(value, index, " " + str(value))

#Plot del grafico della classifica
plt.barh(statistiche.index, statistiche.punti)
plt.show()
```

## Classifica goal fatti

```python
# Trasformo il dizionario ottenuto in un dataframe e lo ordino per punti
statistiche = pd.DataFrame.from_dict(stats).transpose().sort_values("goal_fatti", ascending=True)

#Insert lables
for index, value in enumerate(statistiche.goal_fatti):
    plt.text(value, index, " " + str(value))

#Plot del grafico della classifica
plt.barh(statistiche.index, statistiche.goal_fatti)
plt.show()
```

## Classifica goal subiti

```python
# Trasformo il dizionario ottenuto in un dataframe e lo ordino per punti
statistiche = pd.DataFrame.from_dict(stats).transpose().sort_values("goal_subiti", ascending=False)

#Insert lables
for index, value in enumerate(statistiche.goal_subiti):
    plt.text(value, index, " " + str(value))

#Plot del grafico della classifica
plt.barh(statistiche.index, statistiche.goal_subiti)
plt.show()
```

## Classifica vittorie con pareggi

```python
# Trasformo il dizionario ottenuto in un dataframe e lo ordino per punti
statistiche = pd.DataFrame.from_dict(stats).transpose().sort_values("vittorie", ascending=True)

#Insert lables
for index, value in enumerate(statistiche.vittorie):
    plt.text(value, index, " " + str(value))

#Plot del grafico della classifica
plt.barh(statistiche.index, statistiche.vittorie)
plt.barh(statistiche.index, statistiche.pareggi)

plt.show()
```
